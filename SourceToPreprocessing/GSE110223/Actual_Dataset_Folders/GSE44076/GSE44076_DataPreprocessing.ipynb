# Download Meta data


```python
import pandas as pd
import gzip
import urllib.request

# Define the metadata URL
url = "https://ftp.ncbi.nlm.nih.gov/geo/series/GSE44nnn/GSE44076/matrix/GSE44076_series_matrix.txt.gz"

# Download and read the matrix file
with urllib.request.urlopen(url) as response:
    with gzip.open(response, 'rt') as f:
        lines = f.readlines()

# Extract metadata lines
meta_lines = [line for line in lines if line.startswith("!Sample")]

# Convert metadata to DataFrame
meta_dict = {}
for line in meta_lines:
    parts = line.strip().split("\t")
    key = parts[0].replace("!Sample_", "")
    values = parts[1:]
    meta_dict[key] = values

# Transpose and create DataFrame
meta_df = pd.DataFrame(meta_dict)
meta_df.index.name = "Sample"
meta_df.reset_index(inplace=True)

# Save to CSV
meta_df.to_csv("GSE44076_metadata.csv", index=False)
print("Metadata saved to GSE44076_metadata.csv")
```
