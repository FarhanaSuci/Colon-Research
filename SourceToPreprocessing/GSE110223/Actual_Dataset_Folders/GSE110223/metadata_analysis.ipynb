```python
# List of columns to remove
columns_to_remove = [
    "title","status", "submission_date", "last_update_date", "type", "channel_count",
    'organism_ch1', 'characteristics_ch1', 'molecule_ch1',
       'extract_protocol_ch1', 'label_ch1', 'label_protocol_ch1', 'taxid_ch1',
       'hyb_protocol', 'scan_protocol', 'description', 'data_processing',
       'platform_id', 'contact_name', 'contact_email', 'contact_phone',
       'contact_laboratory', 'contact_department', 'contact_institute',
       'contact_address', 'contact_city', 'contact_state',
       'contact_zip/postal_code', 'contact_country', 'supplementary_file',
       'data_row_count'
]

metadata_cleaned = metadata.drop(columns=columns_to_remove, errors="ignore")
```

```python
metadata_cleaned["target"] = metadata_cleaned["source_name_ch1"].str.extract(r"(normal|colorectal)")
```

```python
metadata_final = metadata_cleaned.drop(columns='source_name_ch1', errors="ignore")
```

```python
metadata_final.head()
```

```python
# Remove quotes from the geo_accession column (it's the first column)
metadata_final.iloc[:, 0] = metadata_final.iloc[:, 0].str.replace('"', '', regex=False)
```

```python
metadata_cleaned= metadata_final
metadata_cleaned.head()
```

```python
# Save the cleaned metadata
metadata_cleaned.to_csv("metadata_cleaned.csv", index=False)

print("Columns removed and saved as metadata_cleaned.csv")
```
